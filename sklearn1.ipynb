```python
#loading the csv data
import pandas as pd
import numpy as np
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
#to split the table into input and output
X = heart_disease.drop("target", axis=1)
y = heart_disease["target"]
```

```python
#importing the model and initializing it
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

#to print the hyperparams of the model
clf.get_params()
```

```python
# Split the data into training and test sets
from sklearn.model_selection import train_test_split

# Example use case (requires X & y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#passing the training data to model
clf.fit(X_train, y_train)
```

```python
#make a prediction
y_preds = clf.predict(X_test)
y_preds
```

```python
#evaluate the model
clf.score(X_train, y_train)
```

```python
#evaluate the model
clf.score(X_test, y_test)
```

```python
#to print some classification metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
# improving the model by tunig the hyper parameter - the n_estimators
# Try different amount of n_estimators
np.random.seed(42)
for i in range(100, 200, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
    print("")
```

```python
# saving the model and reloading it
import pickle

pickle.dump(clf, open("random_forest_model_1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pkl", "rb"))
loaded_model.score(X_test, y_test)
```
