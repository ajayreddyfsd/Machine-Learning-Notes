```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load the heart disease dataset (replace filename if needed)
df = pd.read_csv('./data/heart-disease.csv')

# Separate features (X) and target (y)
# 'target' should be the column with 0 (no disease) or 1 (disease)
X = df.drop('target', axis=1)  # All columns except 'target'
y = df['target']              # The 'target' column is the label

# Split the dataset into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)

# Create the Random Forest classifier
model = RandomForestClassifier(n_estimators=100)

# Train the model using training data
model.fit(X_train, y_train)
```

```python
# making predictions
y_preds = model.predict(X_test)
y_preds
```

```python
# making predictions
# predict_proba is only for classification based models
y_preds = model.predict_proba(X_test)
y_preds
```
