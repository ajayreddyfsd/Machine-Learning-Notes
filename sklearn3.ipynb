```python
#importing required libraries
import numpy as np
import pandas as pd
```

```python
#loading the data
car_sales_missing = pd.read_csv("data/car-sales-extended-missing-data.csv")
car_sales_missing
```

```python
#shows how many missing values in each column
car_sales_missing.isna().sum()
```

```python
# Fill missing values in multiple columns at once using a dictionary
car_sales_missing.fillna({

    # For the 'Make' column, replace NaNs with the string "missing"
    "Make": "missing",

    # For the 'Colour' column, replace NaNs with the string "missing"
    "Colour": "missing",

    # For the 'Odometer (KM)' column, replace NaNs with the mean of the column
    "Odometer (KM)": car_sales_missing["Odometer (KM)"].mean(),

    # For the 'Doors' column, replace NaNs with the number 4
    "Doors": 4,

    # For the 'Price' column, replace NaNs with the number 0
    "Price": 0

# Apply these replacements to the original DataFrame itself (in-place)
}, inplace=True)
```

```python
#shows how many missing values in each column
car_sales_missing.isna().sum()
```

```python
# Split into X/y
X = car_sales_missing.drop("Price", axis=1)
y = car_sales_missing["Price"]
```

```python
# Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                   one_hot,
                                   categorical_features)],
                                   remainder="passthrough")

#passing the X dat
transformed_X = transformer.fit_transform(X)
transformed_X = pd.DataFrame(transformed_X)
transformed_X
```
